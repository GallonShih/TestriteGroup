```python
for i in site:
    os.chdir('C:/Users/RT009223/TestriteGroup\滲透率專案/01_raw data/TLW社區店/'+i)
    files=os.listdir('C:/Users/RT009223/TestriteGroup\滲透率專案/01_raw data/TLW社區店/'+i)
    start=1
    if len(files)>12:
        files=files[:12]
    for j in files:
        print(i,start,j)
        start+=1
```

```python
df01=pd.DataFrame(columns=['id','type','time','site','timeline'])
df02=pd.DataFrame(columns=['id','type','time','site','timeline'])
for i in site:
    os.chdir('C:/Users/RT009223/TestriteGroup/滲透率專案/01_raw data/TLW社區店/'+i)
    files=os.listdir('C:/Users/RT009223/TestriteGroup/滲透率專案/01_raw data/TLW社區店/'+i)
    if len(files)>12:
        files=files[:12]
    start=1
    for j in files:
        df01_2=pd.read_csv(j+'/一般卡會員活躍程度.csv',header=None,names=['id','type'])
        df01_2['time']=j
        df01_2['site']=i
        df01_2['timeline']=start
        df01_3=df01_2[-(df01_2.id.isin(df02[df02['site']==i]['id']))]
        df01=pd.concat([df01,df01_2]).reset_index(drop=True)
        df02=pd.concat([df02,df01_3]).reset_index(drop=True)
        start+=1
    print(df01.shape)
    print(df02.shape)
```

```python
print(df01.groupby(['site'])['id'].count())
print('去重',df02.groupby(['site'])['id'].count())
```

```python
df01_01=df01.groupby(['site','type','time','timeline'])['id'].count().reset_index(name='n_month')
df01_02=df01.groupby(['site','time','timeline'])['id'].count().reset_index(name='n_month_all')
df01_01=pd.merge(df01_01,df01_02)
df01_01['percent_month']=df01_01.n_month/df01_01.n_month_all

#df01_02=pd.pivot_table(df01_01,index=['site','type'],values=['percent'],columns=['time'])
#df01_02.columns=df01_01.time.unique()
#df01_02=df01_02.reset_index()
df01_02=df01_01.copy()
df01_02.type=df01_02.type.apply(lambda x:str(x))
df01_02.type=df01_02.type.replace('1','type1:new')
df01_02.type=df01_02.type.replace('2','type2:active')
df01_02.type=df01_02.type.replace('3','type3:1yr~3yr')
df01_02.type=df01_02.type.replace('4','type4:3yr~')
```

```python
df02_01=df02.groupby(['site','type','time','timeline'])['id'].count().reset_index(name='n_accum')
df02_01['n_accum']=df02_01.groupby(['site','type'])['n_accum'].cumsum()
df02_02=df02.groupby(['site','time','timeline'])['id'].count().reset_index(name='n_accum_all')
df02_02['n_accum_all']=df02_02.groupby(['site'])['n_accum_all'].cumsum()
df02_01=pd.merge(df02_01,df02_02)
df02_01['percent_accum']=df02_01.n_accum/df02_01.n_accum_all

#df02_02=pd.pivot_table(df02_01,index=['site','type'],values=['percent'],columns=['time'])
#df02_02.columns=df02_01.time.unique()
#df02_02=df02_02.reset_index()
df02_02=df02_01.copy()
df02_02.type=df02_02.type.apply(lambda x:str(x))
df02_02.type=df02_02.type.replace('1','type1:new')
df02_02.type=df02_02.type.replace('2','type2:active')
df02_02.type=df02_02.type.replace('3','type3:1yr~3yr')
df02_02.type=df02_02.type.replace('4','type4:3yr~')
```
